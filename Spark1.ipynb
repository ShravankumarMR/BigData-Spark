```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.appName("Top City wise sales").getOrCreate()
```

```python
sc = spark.sparkContext
```

## Data is provided in format :

1)txn id
2)dt
3)cust id
4)amt
5)category
6)prod
7)city
8)state
9)mode of payment

```python
rdd1 = sc.textFile("/content/Input/trans.txt")
print(" Total records : " ,rdd1.count() , "\n\n Sample data: \n")

for i in rdd1.take(5):
  print(i)
```

```python
# check number or partitions created and parallelism supported
print("Number of partitions of data: " ,rdd1.getNumPartitions())
print("Paralleism supported:" , sc.defaultParallelism)
```

```python
# Data is seperated by comma, so split based on comma
rdd2 = rdd1.map(lambda a : (a.split(',')[6], float(a.split(',')[3])))
for i in rdd2.take(5):
  print(i)
```

```python
rdd3 = rdd2.reduceByKey(lambda a,b : a+b)

for i in rdd3.take(5):
  print(i)
```

```python
rdd4 = rdd3.sortBy(lambda a : -a[1])
print("Top City wise sales, for top 10 cities \n")
for i in rdd4.take(10):
  print(i)

print("\n Data is still in these many partitions: " ,rdd4.getNumPartitions())
```

```python
# Get into 1 partition
rdd5 = rdd4.coalesce(1)
```

```python
# save data into output file
rdd5.saveAsTextFile("/content/Output")
```

```python
# Print saves output for demo purpose
rdd6 = sc.textFile("/content/Output/part-00000")
for i in rdd6.take(10):
  print(i)
```
